```python
import pandas as pd

df = pd.read_csv('Investment on education data.csv', skiprows=4)
df.head()
```

```python
df = df.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis=1)
df.head()
```

```python
colleges = pd.read_csv('finisheddataset.csv')

countries = colleges['Country'].unique()

count = colleges['Country'].value_counts().nlargest(30).index

count
```

```python
for country in count:
    if country in df['Country Name'].values:
        print("yes", sep = " ")
    else:
        print(country)
```
