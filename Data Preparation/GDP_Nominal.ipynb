```python
new = pd.DataFrame(columns = columns)
new
```

```python
years = []
for country in list(index):
    data ={}
    for key,value in zip(df[df['Country Name']==country]['Year'].values, df[df['Country Name']== country]['Value'].values):
        data[str(key)] = value
    years.append(data)
```

```python
for row in years:
    new = new.append(row, ignore_index=True)
```
