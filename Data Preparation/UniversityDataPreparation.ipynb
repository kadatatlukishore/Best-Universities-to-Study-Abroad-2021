```python
# We are doing with 2020 dataset



import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
colleges_data = pd.read_csv('projectdata.csv')
colleges_data.set_index('UniversityName', inplace = True)
colleges_data.tail()
```

```python
colleges_data.info()
```

```python
colleges_data.isna().sum()
```

```python
colleges_data['UniversityRank'] = [i+1 for i in range(colleges_data.shape[0])]
```

```python
colleges_data.tail()
```

```python
colleges_data['Total_Num_of_Students'] = [x.replace(',','') for x in colleges_data['Total_Num_of_Students']]
```

```python
colleges_data['Total_Num_of_Students'] = colleges_data['Total_Num_of_Students'].astype('int64')
```

```python
colleges_data.info()
```

```python
colleges_data[colleges_data['UniversityRank']==882]
```

```python
colleges_data['Num_of_Intl_students'] = [x.replace('%','') for x in colleges_data['Num_of_Intl_students']]
colleges_data['Num_of_Intl_students'] = pd.to_numeric(colleges_data['Num_of_Intl_students'])
colleges_data[colleges_data['Num_of_Intl_students'].isna()]
```

```python
colleges_data['Num_of_Intl_students'].fillna(0, inplace=True)
```

```python
colleges_data['Num_of_Intl_students'] = [int(x*(y/100)) for x,y in zip(colleges_data['Total_Num_of_Students'],colleges_data['Num_of_Intl_students'])]
colleges_data.head()
```

```python
colleges_data['Num_of_Intl_students'].isna().sum()
```

```python
# Let's now fill the missing values in femaletomaleratio 

colleges_data['FemaletoMaleRatio'].fillna('1:1', inplace = True)

```

```python
[x for x in colleges_data['FemaletoMaleRatio'] if ': 0' in x]
```

```python
[x for x in colleges_data['FemaletoMaleRatio'] if '00 :' in x]
```

```python
# We will create a new column as MALE TO FEMALE RATIO

colleges_data['maletofemaleratio'] = [int(x.split(':')[1])/int(x.split(':')[0]) for x in colleges_data['FemaletoMaleRatio']]

colleges_data = colleges_data.drop('FemaletoMaleRatio', axis = 1)

colleges_data.head()
```

```python
colleges_data['scores_overall'] = [((float(x[:4])+ float(x[5:]))/2) if (len(x)>4) else float(x) for x in colleges_data['scores_overall']]
colleges_data['scores_overall'].dtypes
```
